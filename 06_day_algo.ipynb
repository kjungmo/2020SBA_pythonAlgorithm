###  합병 정렬 (merge sort)

```python
def partition(seq, start, end):
    pivot = seq[start]
    left = start + 1
    right = end
    done = False
    while not done:
        while left <= right and seq[left] <= pivot:
            left +=1
        while left <= right and pivot < seq[right]:
            right -= 1
            
        if right < left :
            done = True
        else:
            seq[left], seq[right] = seq[right], seq[left]
            
    seq[start], seq[right] = seq[right], seq[start]
    #print(right, seq)
    return right

def quick_sort(seq, start, end):
    if start<end:
        pivot = partition(seq, start, end)
        quick_sort(seq, start, pivot-1)
        quick_sort(seq, pivot+1, end)
    return seq

def test_quick_sort():
    seq = [3, 5, 2, 6, 8, 1, 0, 3, 5, 6, 2]
    seq = quick_sort(seq, 0, len(seq)-1)
    print(seq)
        
if __name__=="__main__":
    test_quick_sort()
```

```python
def partition(seq, start, end):
    pivot = seq[start] #피벗이니까
    left = start + 1 # 그 다음 숫자부터
    right = end
    done = False # 끝났을 때를 알려주는 flag값 
    while not done:     #done이 False니까 이게 true가 될때 까지 돌리자
        while left <= right and seq[left] <= pivot: # left는 pivot보다 큰 값을 찾아다님
            left +=1
        while left <= right and pivot < seq[right]: # right는 pivot보다 작은값을 찾아다님
            right -= 1
#         전부 left와 right가 엇갈리지 않을 때까지 while 돈다는 것이다.
#         피벗의 left에서 피벗은 3보다 작거나 같은 값을 찾아야 한다. 
#     찾을 때까지 left는 피벗보다 작거나 같은 값을 찾으러 계속 이동한다
#     반대로 피벗보다 오른쪽이 큰값을 찾아야 한다
#     2 라 멈춤
#     값을 찾은 경우 if 오른쪽<왼쪽이라면 done을 True로 바꾸고 종료
            
        if right < left : # 엇갈려있니? 
            done = True
        else:
            seq[left], seq[right] = seq[right], seq[left] # 스왑해주는 것
    
    seq[start], seq[right] = seq[right], seq[start]
    #print(right, seq)
    return right

# 결국 pivot인 3보다 작은 것은 왼쪽에, 큰 값은 오른쪽에 소팅해줌
# 그다음에 나눈 곳으로 3이 스왑하게 됨
# 기준점을 기준으로 작은값, 큰값 분할을 한 뒤 계속 그 분할을 이어나감
# 왼쪽의 다섯개를 다시 partition하고 오른쪽 다섯개도partition하면 된다
# 그래서 그것을 재귀적으로 호출하면 pivot자리를 딱 리턴하게 된다
# pivot 기준으로 계속해서 재귀 호출하면 된다.
# 호출한 partition을 가지고 계속 왼쪽과 오른쪽을 호출한다. 

def quick_sort(seq, start, end):
    if start<end: # 원소가 두개이상일때 까지 계속 돌아라. 이것 자체가 종료조건
        pivot = partition(seq, start, end)
        quick_sort(seq, start, pivot-1) # 왼쪽 배열
        quick_sort(seq, pivot+1, end)  # 오른쪽 배열
#         이들을 중심으로 계속해서 partition한다
    return seq

def test_quick_sort():
    seq = [3, 5, 2, 6, 8, 1, 0, 3, 5, 6, 2]
    seq = quick_sort(seq, 0, len(seq)-1) #배열의 첫 인덱스 ~ 마지막 인덱스
    print(seq)
        
if __name__=="__main__":
    test_quick_sort()
```

### 힙 정렬(heap sort)


```python
def heap_sort3(seq):
        for start in range((len(seq)-2)//2, -1, -1):
            siftdown(seq, start, len(seq)-1)
        for end in range(len(seq)-1, 0, -1):
            seq[end], seq[0] = seq[0], seq[end]
            siftdown(seq, 0, end-1)
        return seq
    
def siftdown(seq, start, end):
    root = start
    while True:
        child = root * 2 + 1
        if child > end:
            break
        if child + 1<= end and seq[child] < seq[child+1]:
            child += 1
        if seq[root] < seq[child]:
            seq[root], seq[child] = seq[child], seq[root]
            root = child
        else:
            break
            
def test_heap_sort():
    seq = [3, 5, 2, 6, 8, 1, 0, 3, 5, 6, 2]
    seq = heap_sort3(seq)
    print(seq)
    
if __name__=="__main__":
    test_heap_sort()
```

```python
# 힙을 만들 때에는 트리구조를 만든다
# 출발하자마자 처음의 for문을 때린다.
# ((11-2)//2라 4이고 index 4에서 start한다. 4부터 0까지 돈다, 
# len(seq)-1//2 는 마지막 부모 인덱스이다
# end는 현재 seq의 마지막이 전달되기 때문에 마지막 index가 된다
# root를 start로 본다면 index 4값이 된다. 
# root가 지금은 값은 8인데 루트의 index로 본다면 4이다. 
# start가 4라면 child가 9이된다. 
# child는 11이 된다면 범위를 넘어간 느낌이 든다
# child가 end범위를 넘어간게 된다면 결국 break된다 siftdown 함수가 종료된다
# 뭔가 반복을 통해서 heap을 만들어간다는게 보인다
# 두번째 라운딩에서는 start가 index로는 3가 된다. end는 똑같다.
# root 가 3이고  child가 7가 된다
# 이 경우 break를 타지 않는다
# 밑에 조건을 돌리게 되는데
# 7+1이 8이어서 index 7와 index 8을 비교한다. 3과 5를 비교해서 만약 index 7이 작다면(위배)
# 그래서 if문(왼쪽과 오른쪽 중에 더 큰값을 갖게 하는 역할을 함) 빠져나옴
# root와 child 값에서 root가 더 작다면 child와 값을 스왑
# 이렇게 하고 끝난다. 

def heap_sort3(seq):
        for start in range((len(seq)-2)//2, -1, -1):
            siftdown(seq, start, len(seq)-1) # 여기까지가 힙을 구성하게 하는 코드
        for end in range(len(seq)-1, 0, -1):
            seq[end], seq[0] = seq[0], seq[end]
            siftdown(seq, 0, end-1)
        return seq
    
def siftdown(seq, start, end):
    root = start
    while True:
        child = root * 2 + 1
        if child > end:
            break
        if child + 1<= end and seq[child] < seq[child+1]:
            child += 1
        if seq[root] < seq[child]:
            seq[root], seq[child] = seq[child], seq[root]
            root = child
        else:
            break
            
def test_heap_sort():
    seq = [3, 5, 2, 6, 8, 1, 0, 3, 5, 6, 2]
    seq = heap_sort3(seq)
    print(seq)
    
if __name__=="__main__":
    test_heap_sort()
```

### 이진 검색 

```python
def binary_search_rec(seq, target, low, high):
    if low> high:
        return None
    mid = (low + high) //2
    if target == seq[mid]:
        return mid
    elif target < seq[mid]:
        return binary_search_rec(seq, target, low, mid-1)
    else:
        return binary_search_rec(seq, target, mid+1, high)

def binary_search_iter(seq, target):
    high, low = len(seq), 0
    while low<high:
        mid = (high + low) //2
        if target == seq[mid]:
            return mid
        elif target<seq[mid]:
            high = mid
        else:
            low = mid + 1
    return None

def test_binary_search():
    seq = [1, 2, 5, 6, 7, 10, 12, 14, 15]
    target = 6
    print(binary_search_iter(seq, target))
    print(binary_search_rec(seq, target, 0, len(seq))) # 인덱스가 출력된다.



if __name__=="__main__":
    test_binary_search()
```
